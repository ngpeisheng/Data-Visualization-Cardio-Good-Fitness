```python
#putting alpha code 2 into a list and putting alpha code 3 into another list.

ac2list = [i.alpha_2 for i in list(pycountry.countries)]
ac3list = [i.alpha_3 for i in list(pycountry.countries)]

def country_name(df_hb):
    if(len(df_hb['country'])==2 and df_hb['country'] in ac2list):
        return pycountry.countries.get(alpha_2=df_hb['country']).name
    elif(len(df_hb['country'])==3 and df_hb['country'] in ac3list):
        return pycountry.countries.get(alpha_3=df_hb['country']).name
    else:
        return 'East Timor'
    
df_hb['country'] = df_hb.apply(country_name, axis = 1)
```
